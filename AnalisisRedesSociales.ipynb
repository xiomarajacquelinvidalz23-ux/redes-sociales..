```python
import unicodedata
import pandas as pd

# Función para normalizar texto (minúsculas, quitar acentos, espacios)
def normalizar_texto(s):
    if pd.isnull(s):
        return s
    s = str(s).casefold()  # minúsculas
    s = ''.join(c for c in unicodedata.normalize('NFKD', s) if not unicodedata.combining(c))  # quitar acentos
    s = s.strip()
    return s

# Normalizar la columna de ocupación
df_redesSociales['Ocupacion_Normalizada'] = df_redesSociales['Ocupacion_Principal_ES'].apply(normalizar_texto)

# Diccionario de equivalencias para estudiantes
mapa = {
    'estudante': 'Estudiante',
    'estudiante': 'Estudiante',
    'alumno': 'Estudiante',
    'universitario': 'Estudiante'
}

# Reemplazar valores normalizados por categorías limpias
df_redesSociales['Ocupacion_Normalizada'] = df_redesSociales['Ocupacion_Normalizada'].replace(mapa)

# Contar estudiantes
num_estudiantes = (df_redesSociales['Ocupacion_Normalizada'] == 'Estudiante').sum()
print(f"Número de personas que son estudiantes: {num_estudiantes}")
```
